## Functions

```python
# Function Definitions
# Curve fitting part
# define the true objective function ----------------------------------------------
def psyFunction(x,mu,sd,k,offset): #Sigmoid function
    yhat = norm.cdf(x,mu, sd) * k + offset
    return yhat
def Create2DList(rows,cols,initVal):
    answrs=[]
    for j in range(rows):
        column = []
        for i in range(cols):
            column.append(initVal)
        answrs.append(column)
    return answrs
def Average(lst):

    # Make sure no number 2s are included in the average
    if 2 in lst:
        lst.remove(2.0)

    avrg = 0.0
    try:
        avrg = np.round(sum(lst) / len(lst),3)
    except Exception as e:
#         print(e)
        avrg = np.nan

    return avrg
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx, array[idx]

def r2(x, y):
    corr = stats.pearsonr(x, y) #[0] ** 2
    return corr
```

## Normal distribution check

```python
# Visual 
plt.figure()
mask = (df8['Experiment'] == 'Vis')
sns.histplot(x='JND', data = df8[mask], kde=True)
alpha = 1e-3

k2, p = stats.normaltest(df8[mask]['JND'])

print("p = {:g}".format(p))

if p < alpha:  # null hypothesis: x comes from a normal distribution
    print("Data is normally distributed")
else:
    print("The null hypothesis cannot be rejected")
    
    
# Visual-proprioceptive  
plt.figure()
mask = (df8['Experiment'] == 'Vip')
sns.histplot(x='JND', data = df8[mask], kde=True)
alpha = 1e-3

k2, p = stats.normaltest(df8[mask]['JND'])

print("p = {:g}".format(p))

if p < alpha:  # null hypothesis: x comes from a normal distribution
    print("Data is normally distributed")
else:
    print("The null hypothesis cannot be rejected")
```

```python
# Visual 
plt.figure()
mask = (df8['Experiment'] == 'Vis')
sns.histplot(x='Bias', data = df8[mask], kde=True)
alpha = 1e-3

k2, p = stats.normaltest(df8[mask]['Bias'])

print("p = {:g}".format(p))

if p < alpha:  # null hypothesis: x comes from a normal distribution
    print("Data is normally distributed")
else:
    print("The null hypothesis cannot be rejected")
    
# Visual-proprioceptive
plt.figure()
mask = (df8['Experiment'] == 'Vip')
sns.histplot(x='Bias', data = df8[mask], kde=True)
alpha = 1e-3

k2, p = stats.normaltest(df8[mask]['Bias'])

print("p = {:g}".format(p))

if p < alpha:  # null hypothesis: x comes from a normal distribution
    print("Data is normally distributed")
else:
    print("The null hypothesis cannot be rejected")
```
